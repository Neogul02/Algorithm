# 객체지향 프로그래밍 (OOP)
> OOP는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">데이터(속성)와 기능(메서드)을 하나의 객체로 묶어</mark> 프로그램을 구조화하는 패러다임입니다.

>절차적 프로그래밍에서는 데이터(변수)와 동작(함수)이 분리되어 있습니다. 학생 정보를 처리하는 코드가 100줄이 되면, 어떤 함수가 어떤 변수를 사용하는지 추적하기 어려워집니다.</br>
프로그램 규모가 커질수록 관련 없는 함수가 중요한 변수를 실수로 수정하는 문제가 발생합니다.</br>
>OOP는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">관련된 데이터와 기능을 하나의 "객체"로 묶어서</mark> 이런 복잡성을 관리하며, 현실 세계의 사물을 모델링하듯이 코드를 구조화할 수 있어 대규모 소프트웨어 개발에 적합합니다.

><mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">클래스</mark>는 설계도입니다. "학생"이라는 클래스는 이름, 나이, 점수를 가지고 자기소개를 할 수 있다고 정의합니다.</br>
><mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">인스턴스(객체)</mark>는 설계도로 만든 실제 물건입니다. `Student("철수", 20)`은 철수라는 구체적인 학생입니다.</br>
>같은 클래스로 수많은 인스턴스를 만들 수 있습니다. 설계도는 하나지만 집은 여러 채 지을 수 있는 것과 같습니다.</br>
>클래스를 배우면 Python의 모든 것이 왜 "객체"인지, `str`, `list`, `dict`가 왜 메서드를 가지는지 이해하게 됩니다.

## 클래스 정의 (Class Definition)
> `class` 키워드로 클래스를 정의하고, `__init__`으로 초기화합니다.

```python
# TODO 1: name, age를 속성으로 갖고 introduce() 메서드가 자기소개 문자열을 반환하는 Student 클래스를 정의하고, 이름은 '철수', 나이는 '20'인 student1 객체를 생성하여 introduce()를 호출해봅시다.
class Student:
    
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def introduce(self):
        return f'제 이름은 {self.name}이고 {self.age}살 입니다.'
    
student1 = Student('철수','20')
print(student1.introduce())
```

💡self의 의미
> `self`는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">현재 인스턴스 자신</mark>을 가리킵니다.</br>
> Java의 `this`와 같은 역할이지만, Python에서는 명시적으로 첫 번째 매개변수로 작성해야 합니다.

## 매직 메서드 (Magic Methods)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">더블 언더스코어(__)</mark>로 감싼 특수 메서드로, Python의 내장 동작을 커스터마이즈합니다.

<table style="width:100%">
  <thead>
    <tr>

    </tr>
  </thead>
  <tbody>
    <tr><td style="text-align:center"><code>__init__</code></td><td>인스턴스 생성 시</td><td>초기화 (생성자)</td></tr>
    <tr><td style="text-align:center"><code>__str__</code></td><td><code>print()</code>, <code>str()</code> 호출 시</td><td>사용자 친화적 문자열</td></tr>
    <tr><td style="text-align:center"><code>__repr__</code></td><td>인터프리터에서 출력 시</td><td>개발자용 문자열</td></tr>

  </tbody>

```python
# TODO 2: x, y를 속성으로 갖고 __str__()로 "Vector(x, y)"를 반환하며, __add__()로 두 벡터를 더하는 Vector 클래스를 정의하고, Vector(1, 2)와 Vector(3, 4)를 생성하여 출력과 덧셈 결과를 확인해봅시다.

class Vector:
    
    def __init__(self, x, y):
        self.x = x;
        self.y = y;
        
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f'Vector({self.x}, {self.y})'
    
vector1 = Vector(1, 2)
vector2 = Vector(3, 4)

print(vector1)
print(vector2)
print(vector1 + vector2)
        
    
```

### __repr__, __len__, __eq__, __lt__ 실습
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">비교 및 크기 연산</mark>을 위한 매직 메서드를 구현해봅시다.

```python
# TODO 3: label과 items(리스트)를 속성으로 갖는 Box 클래스를 정의하고, __repr__()로 "Box(label=..., items=...)" 형식을, __len__()으로 items의 길이를,
# __eq__()로 items 비교를, __lt__()로 items 길이 비교를 구현한 뒤, Box("A", [1,2,3])와 Box("B", [1,2,3])와 Box("C", [1,2])를 생성하여 repr, len, ==, < 연산 결과를 출력해봅시다.
class Box:
    
    def __init__(self, label:str, items:list):
        self.label = label
        self.items = items


box1 = Box("A",[1,2,3])
box2 = Box("B", [1,2,3])
```

💡__str__ vs __repr__
> `__str__`은 사용자가 읽기 좋은 문자열, `__repr__`은 디버깅용으로 객체를 재현할 수 있는 문자열을 반환합니다.</br>
> `__repr__`만 정의하면 `__str__` 미정의 시 `__repr__`이 대신 사용됩니다.

## 상속 (Inheritance)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">기존 클래스의 속성과 메서드를 물려받아</mark> 새로운 클래스를 만드는 기법입니다.

```python
# TODO 4: name을 속성으로 갖고 speak()가 "..."을 반환하는 Animal 클래스를 정의하고, 이를 상속받아 speak()를 "멍멍!"으로 오버라이딩하는 Dog, "야옹!"으로 오버라이딩하는 Cat 클래스를 만든 후, Dog("바둑이")와 Cat("나비")를 생성하여 출력해봅시다.
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return '...' # 말 못함
    
# 상속
class Dog(Animal):
    
    def speak(self):
        return '멍멍!'
    
class Cat(Animal):
    
    def speak(self):
        return '니으으야오오옹!'

gae = Dog("요태리");
nyang = Cat("썌비냥");

print(gae.name, gae.speak() );
print(nyang.name, nyang.speak() );   
```

```python
# TODO 5: Dog을 상속받는 GuideDog 클래스를 정의하고, 부모 클래스를 초기화한 뒤 owner 속성을 추가한 후, GuideDog("진돌이", "철수")를 guide_dog에 생성하여 name과 owner를 출력해봅시다.
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return '...' # 말 못함
    
# 상속
class Dog(Animal):
    
    def speak(self):
        return '멍멍!'

class GuideDog(Dog):
    def __init__(self, name, owner):
        super.__init__(name)
        self.owner = owner
        
guide_dog = GuideDog("땡칠이","김강철");
print(guide_dog.name, guide_dog.owner)
```

💡super()
> `super()`를 사용하면 부모 클래스의 메서드를 호출할 수 있습니다.

```python
# TODO 6: 클래스 변수 count=0을 갖는 Counter 클래스를 정의하고, __init__에서 Counter.count를 1 증가시킨 뒤 self.id에 저장하게 만든 후, Counter()를 두 번 생성하여 Counter.count와 각 인스턴스의 id를 출력해봅시다.
class Counter:
    count = 0
    
    def __init__(self):
        Counter.count+=1
        self.id = Counter.count
        
counter1 = Counter();
counter2 = Counter();

print(Counter.count, counter1.id, counter2.id)
```
