# 기본 자료형
> Python은 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">동적 타이핑(Dynamic Typing)</mark> 언어로, 변수 선언 시 타입을 명시하지 않습니다.</br>
> 값이 할당되는 시점에 타입이 자동으로 결정됩니다.

>자료형은 프로그램이 다루는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">데이터의 종류와 허용 연산을 정의</mark>하는 가장 기본적인 개념입니다.</br>
>컴퓨터는 숫자와 문자를 다르게 저장하기 때문에, `"3" + "5"`는 `"35"`이고 `3 + 5`는 `8`입니다.</br>
>자료형을 모르면 이런 차이를 예측하지 못해 버그를 만들게 됩니다.

>Java나 C에서는 변수를 선언할 때 타입을 직접 명시해야 합니다: `int x = 5;`</br>
>반면 Python은 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">동적 타이핑</mark> 언어로, 값을 할당하는 순간 타입이 자동으로 결정됩니다: `x = 5`</br>
>같은 변수에 다른 타입의 값을 나중에 재할당하는 것도 허용됩니다.</br>
>이 덕분에 코드가 짧고 유연하지만, 타입 오류를 실행 전에 알기 어렵다는 단점도 있습니다.

## 수치형 내장 함수 (Numeric)
> 수치형 데이터를 다루는 내장 함수입니다.

```python
# TODO 4: -3.14의 절대값을 출력해봅시다.
print(abs(-3.14))
```

```python
# TODO 5: 3.7의 반올림 결과를 출력해봅시다.
print(round(3.7))
```

```python
# TODO 6: 2의 3제곱을 출력해봅시다.
print(2**3)
```

```python
# TODO 7: 0.1 + 0.2와 0.3이 같은지 출력해봅시다.
print(True if ((0.1+0.2) == 0.3) else False)
```

```python
# TODO 8: math 모듈을 import하고, 0.1 + 0.2와 0.3이 근사적으로 같은지 비교한 결과를 출력해봅시다.
import math;
print(math.isclose(0.1+0.2, 0.3))
```

💡math.isclose()
> `rel_tol` 속성을 이용하여 상대 오차 허용 범위를 지정할 수 있습니다.</br>
> 부동소수점 비교 시 `==` 대신 `math.isclose()`를 사용하는 것이 안전합니다.

## 문자열 (String)
> 문자열은 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">불변(Immutable)</mark> 객체로, 메서드 호출 시 새로운 문자열을 반환합니다.

```python
# TODO 9: "Hello" 문자열을 소문자로 변환한 결과를 출력해봅시다.
user_text = 'Hello'
print(user_text.lower())
```

```python
# TODO 10: "Hello" 문자열을 대문자로 변환한 결과를 출력해봅시다.
user_text = 'Hello'
print(user_text.upper())
```

```python
# TODO 11: "test.py" 문자열에서 끝이 ".py"로 끝나는지 확인해봅시다.
user_text = 'test.py'
print(user_text[-3:]== ".py")
```

```python
# TODO 12: "123" 문자열이 숫자로만 이루어져 있는지 출력해봅시다.
user_text = "123"
print(user_text.isdigit())
```

```python
# TODO 13: "a"와 "b"를 연결한 결과를 출력해봅시다.
print('a'+'b')
```

### 추가 문자열 메서드
> 문자열 처리에 자주 사용되는 핵심 메서드입니다.

```python
# TODO 14: "  hello  " 문자열의 앞뒤 공백을 제거한 결과를 출력해봅시다.
user_str = "  hello  "
print(user_str.strip())
```

```python
# TODO 15: "a,b,c" 문자열을 ","로 분리한 결과를 출력해봅시다.
user_str = "a,b,c"
print(user_str.split(","))
```

```python
# TODO 16: " "를 구분자로 ["hello", "world"] 리스트를 하나의 문자열로 결합한 결과를 출력해봅시다.
print(" ".join(["hello", "world"]))
```

```python
# TODO 17: "hello world" 문자열에서 "world"를 "Python"으로 치환한 결과를 출력해봅시다.
user_text = 'hello world'
user_text.replace('world', 'python')
print(user_text)
```

```python
# TODO 18: "hello" 문자열에서 "ll"의 위치(인덱스)를 find()로 출력해봅시다.
print('hello'.find('ll'))
```

```python
# TODO 19: "hello" 문자열이 "he"로 시작하는지 startswith()로 확인하여 출력해봅시다.
print('heelo'.startswith('he'))
```

💡find() vs index()
> `find()`는 찾지 못하면 `-1`을 반환하고, `index()`는 `ValueError`를 발생시킵니다.</br>
> 안전한 검색에는 `find()`를 사용하세요.

## 형 변환 (Type Casting)
> `int()`, `float()`, `str()` 함수로 타입 간 변환이 가능합니다.

```python
# TODO 20: 문자열 "42"를 정수로 변환한 값을 num 변수에 저장하고, num과 num의 타입을 출력해봅시다.
num = int('42')
print(f'num: {num}, numType: {type(num)}')
```

```python
# TODO 21: 정수 10을 실수로 변환한 값을 val 변수에 저장하고, val과 val의 타입을 출력해봅시다.
val = float(10)
print(f'num: {val}, numType: {type(val)}')
```

```python
# TODO 22: 실수 3.14를 문자열로 변환한 값을 text 변수에 저장하고, text와 text의 타입을 출력해봅시다.
text = str(3.14)
print(f'num: {text}, numType: {type(text)}')
```
