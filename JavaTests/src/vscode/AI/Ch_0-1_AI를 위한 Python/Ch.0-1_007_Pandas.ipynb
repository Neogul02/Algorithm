# Pandas 기본 구조
> Pandas는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">표 형태의 데이터를 다루는</mark> Python 라이브러리입니다.

>NumPy는 숫자 연산에 최적화되어 있지만, 현실의 데이터는 훨씬 복잡합니다.</br>
이름, 나이, 도시처럼 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">서로 다른 타입이 섞인 데이터</mark>를 NumPy 배열 하나로 표현하기 어렵고, 열 이름 개념이 없어 `arr[:, 2]`처럼 숫자 인덱스로만 접근해야 합니다.</br>
>결측치(`NaN`) 처리, 날짜 데이터, CSV/Excel 파일 입출력 등의 기능도 NumPy에는 없습니다.

>Pandas는 이런 문제들을 해결합니다. <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">열 이름으로 접근</mark>하여 `df["age"]`처럼 의미 있는 이름으로 데이터를 선택할 수 있고, 숫자, 문자열, 날짜를 하나의 테이블에 담는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">혼합 타입</mark>을 지원합니다.</br>
>`NaN`을 인식하고 `dropna()`, `fillna()`로 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">결측치를 자동 처리</mark>하며, CSV, Excel, JSON, SQL 등을 한두 줄로 읽고 쓸 수 있습니다.</br>
>SQL의 `GROUP BY`처럼 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">강력한 그룹 연산</mark>으로 데이터를 집계하고 분석할 수 있어, Pandas는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">데이터 분석의 엑셀이라고 불릴 만큼</mark> 데이터 탐색과 전처리의 표준 도구입니다.

```python
# TODO 16: name과 score 열로 구성된 두 DataFrame을 pd.concat()으로 연결하고 인덱스를 초기화하여 출력해봅시다.
df_a = pd.DataFrame({
    "name": ["Alice"],
    "score": [90]
})

df_b = pd.DataFrame({
    "name": ["Bob"],
    "score": [96]
})

combined = pd.concat([df_a, df_b], ignore_index=True)
print(combined)
```
