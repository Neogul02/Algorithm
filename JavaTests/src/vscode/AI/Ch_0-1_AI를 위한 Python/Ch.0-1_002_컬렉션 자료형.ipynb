# 컬렉션 자료형 (Collection Types)
> Python의 컬렉션은 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">여러 데이터를 하나의 변수에 저장</mark>하는 자료형입니다.

>학생 30명의 점수를 저장한다고 가정해봅시다. 단일 변수만 사용하면 `score1`, `score2`, ... `score30`처럼 30개의 변수가 필요합니다.</br>
>컬렉션을 사용하면 `scores = [85, 90, 78, ...]` 한 줄로 해결됩니다.</br>
>어떤 컬렉션을 선택하느냐에 따라 코드의 효율과 안전성이 달라집니다.

>각 컬렉션에는 적합한 사용 시나리오가 있습니다. <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">list</mark>는 순서가 있고 변경이 필요한 데이터에, <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">tuple</mark>은 변경되어선 안 되는 데이터에, <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">set</mark>은 중복 제거나 집합 연산이 필요할 때, <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">dict</mark>는 이름으로 데이터를 찾아야 할 때 사용합니다.</br>
>자료구조 선택은 단순한 문법 문제가 아니라 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">성능과 코드 의도를 표현하는 설계 결정</mark>입니다.

```python
# TODO 1: fruits에 1, 2, "apple", True 를 저장하고 출력해봅시다.
fruits = [1, 2, "apple", True]
print(fruits)
```

```python
# TODO 6: fruits의 얕은 복사본을 fruits_copy에 저장하고, fruits_copy에 "mango"를 추가한 뒤 fruits와 fruits_copy를 각각 출력해봅시다.
fruits_copy = fruits.copy()
fruits_copy.append("mango")
print(fruits)
print(fruits_copy)
```

### 인덱싱과 슬라이싱 (Indexing & Slicing)

```python
# TODO 8: numbers에 10, 20, 30, 40, 50을 저장하고, 1번 인덱스부터 2번 인덱스의 슬라이싱 결과를 출력해봅시다.
numbers = [10, 20, 30, 40, 50]
print(numbers[0:2])
```

```python
# TODO 10: numbers의 짝수 인덱스 요소를 출력해봅시다.
print(numbers[::2])
```

### 리스트 컴프리헨션 (List Comprehension)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">간결한 반복 표현</mark>으로 리스트를 생성하는 문법입니다.

```python
# TODO 11: 리스트 컴프리헨션으로 1부터 4의 제곱을 squares에 저장하고 출력해봅시다.
print([i**2 for i in range(1,5)])
```

```python
# TODO 12: 리스트 컴프리헨션으로 1부터 9까지 수 중에서 '짝수'를 evens에 저장하고 출력해봅시다.
evens = [i for i in range(2,10) if i % 2 == 0]
print(evens)
```

## 튜플 (Tuple)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">불변(Immutable)</mark> 순서형 자료구조로, 생성 후 수정할 수 없습니다.

```python
# TODO 13: point에 (3, 4) 튜플을 저장하고, x와 y로 언패킹하여 f-string으로 출력해봅시다.
point = (3, 4)
x, y = point
print(f'x:{x}, y:{y}')
```

```python
# TODO 14: try-except를 사용하여 point의 0번 인덱스에 10을 대입 시 발생하는 TypeError를 확인해봅시다.
point = (3, 4)
try:
    point[0] = 10
except TypeError as e:
    print(f"TypeError 발생: {e}")
```

💡튜플 사용 시점
> 변경되면 안 되는 데이터(좌표, 설정값 등)에 사용합니다.</br>
> 딕셔너리의 키로 사용할 수 있습니다. (리스트는 불가)

## 집합 (Set)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">중복을 허용하지 않는</mark> 비순서형 자료구조입니다.

```python
# TODO 15: set_a에 1, 2, 3, 4 집합을, set_b에 3, 4, 5, 6 집합을 생성하고 f-string으로 출력해봅시다.
set_a = {1,2,3,4}
set_b = {3,4,5,6}

print(set_a)
print(set_b)
```

```python
# TODO 16: 합집합, 교집합, 차집합 연산자를 사용하여 set_a와 set_b의 집합 연산 결과를 출력해봅시다.
print(set_a | set_b) # 합집합
print(set_a & set_b) # 교집합
print(set_a - set_b) # 차집합
```

```python
# TODO 17: number_list 원소의 중복을 제거한 후 리스트로 변환하여 unique 변수에 저장하고 출력해봅시다.
number_list = [1,2,2,3,3,3]
unique = set(number_list)

print(list(unique))
```

## 딕셔너리 (Dictionary)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">키-값(Key-Value) 쌍</mark>으로 데이터를 저장하는 매핑형 자료구조입니다.

```python
# TODO 18: student에 name은 김철수, age는 20, grade는 A인 딕셔너리를 저장하고 출력해봅시다.
student = {'name':'김철수','age':20, 'grade': 'A'}
print(student)
```

```python
# TODO 19: student에 "major" 키로 "CS" 값을 추가하고 출력해봅시다.
student["major"] = "CS"
print(student)
```

```python
# TODO 20: students의 Key와 Values를 각각 출력해봅시다.
print(student.keys())
print(student.values())
```

```python
# TODO 21: student에서 name을 조회하여 값이 있다면 그 값을 없다면 "" (빈값)을, phone을 조회하여 값이 있다면 그 값을 없다면 "없음"을 출력해봅시다.
print(student.get('name',''))
print(student.get('phone','없어용'))
```

💡list vs dict 조회 성능
> 리스트의 `in` 검색은 O(n)이지만, 딕셔너리의 키 검색은 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">O(1)</mark>입니다.</br>
> 대량 데이터 검색 시 딕셔너리가 훨씬 효율적입니다.

### 딕셔너리 추가 메서드 및 컴프리헨션
> 딕셔너리 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">병합과 간결한 생성</mark>을 위한 문법입니다.

```python
# TODO 22: base에 name은 "철수", age는 20을, extra에 city는 "서울", grade는 "A"를 저장하고, base에 extra를 병합한 뒤 출력해봅시다.

base ={'name':'철수', 'age':20}
extra = {'city':'서울', 'grade':'A'}

base.update(extra)
print(base)
```

```python
# TODO 23: 딕셔너리 컴프리헨션으로 1부터 5까지 각 숫자를 키, 그 제곱을 값으로 하는 squares를 만들고 출력해봅시다.
squares = {x: x**2 for x in range(1,6)}
print(squares)
```

```python
# TODO 24: 딕셔너리 컴프리헨션으로 1부터 5까지 수 중 짝수만 키, 그 제곱을 값으로 하는 even_squares를 만들고 출력해봅시다.
squares = {x: x**2 for x in range(1,6) if x%2==0}
print(squares)
```

💡update() vs 직접 할당
> `update()`는 여러 키를 한 번에 추가/수정할 때 유용합니다.</br>
> Python 3.9+에서는 `dict1 | dict2` 연산자로 딕셔너리를 병합할 수 있습니다.
