## 내장 함수 (Built-in Functions)
> Python이 기본 제공하는 고차 함수로, <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">함수형 프로그래밍</mark> 스타일을 지원합니다.

### map() - 함수 매핑
> 이터러블의 각 요소에 함수를 적용하여 새 이터러블을 반환합니다.

```python
# TODO 9: numbers에 [1, 2, 3, 4, 5]를 저장하고, map()과 람다를 사용하여 각 요소를 제곱한 결과를 squared에 리스토로 저장하여 출력해봅시다.
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x : x ** 2, numbers))
print(squared)
```

</br>

### filter() - 조건 필터링
> 이터러블에서 조건을 만족하는 요소만 걸러냅니다.

```python
# TODO 10: numbers에 1, 2, 3, 4, 5, 6을 저장하고, filter()와 람다를 사용하여 짝수만 evens에 리스트로 저장하여 출력해봅시다.
numbers = [1,2,3,4,5,6]
evens = list(filter(lambda x: x%2==0,numbers))
print(evens)
```

</br>

### any() / all() - 논리 검사
> `any()`는 하나라도 참이면 `True`, `all()`은 모두 참이어야 `True`입니다.

```python
# TODO 11: scores에 80, 90, 55, 70을 저장하고, any()로 90 이상인 점수가 하나라도 있는지, all()로 모두 60 이상인지를 각각 출력해봅시다.
scores = [80, 90, 55, 70]

print(any(score >= 90 for score in scores))
print(all(score >= 60 for score in scores))
```

</br>

### sorted() with key - 정렬 기준 지정
> `key` 매개변수로 정렬 기준 함수를 지정합니다.

```python
# TODO 12: students에 ("철수", 85), ("영희", 92), ("민수", 78)을 저장하고, sorted()와 람다를 사용하여 점수 기준 내림차순으로 정렬한 결과를 by_score에 저장하여 출력해봅시다.
students = [("철수", 85), ("영희", 92), ("민수", 78)]
by_score = sorted(students, key=lambda x: x[1])
print(by_score)
```

💡map/filter vs 리스트 컴프리헨션
> `map()`과 `filter()`는 지연 평가(lazy evaluation)로 메모리 효율적입니다.</br>
> 가독성이 중요할 때는 리스트 컴프리헨션, 대용량 데이터에는 `map`/`filter`를 고려하세요.
