# 조건문 (Conditional Statements)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">if/elif/else</mark> 구조로 조건에 따라 실행할 코드 블록을 선택합니다.

>컴퓨터는 기본적으로 코드를 위에서 아래로 순서대로 실행합니다. 하지만 현실의 문제는 항상 "상황에 따라 다르게" 처리해야 합니다.</br>
>로그인할 때 비밀번호가 맞으면 환영 메시지를, 틀리면 오류 메시지를 보여줘야 하고, 100명의 학생 성적을 계산할 때 같은 공식을 100번 반복 적용해야 합니다.</br>
><mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">제어문과 반복문은 이런 "조건 분기"와 "반복 실행"을 가능하게 하는 핵심 도구</mark>이며, 이것 없이는 어떤 실용적인 프로그램도 작성할 수 없습니다.

>Python의 조건문은 `True`/`False`뿐 아니라 다양한 값을 조건으로 받을 수 있습니다. `if lst:`처럼 쓰면 리스트가 비어 있는지 확인할 수 있어 `if lst != []:`보다 훨씬 간결합니다.</br>
><mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">Falsy 값</mark>(`0`, `""`, `[]`, `{}`, `None`)을 모르면 의도치 않은 버그가 생깁니다. 예를 들어 점수가 `0`점인 학생을 `if score:`로 검사하면 0점을 "없는 값"으로 취급하게 됩니다.</br>
>Truthy/Falsy 규칙을 이해하면 더 간결하고 안전한 조건문을 작성할 수 있습니다.

```python
# TODO 1: score = 85를 저장하고, 90 이상이면 "A", 80 이상이면 "B", 70 이상이면 "C", 그 외 "F"를 grade에 저장한 후 f-string으로 출력해봅시다.
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"
    
print(f"점수: {score}, 학점: {grade}")
```

## for 문
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">반복 가능한 객체(Iterable)</mark>를 순회하며 각 요소에 대해 코드를 실행합니다.

```python
# TODO 3: fruits에 "사과", "바나나", "포도" 리스트를 만들고, for문으로 각 요소를 출력해봅시다.
fruits = ["사과", "바나나", "포도"]
for fruit in fruits:
    print(fruit)
```

```python
# TODO 4: range(2, 10, 3)을 사용하여 for문으로 반복하며 각 값을 출력해봅시다. (print의 end=" " 옵션 사용)
for i in range(2, 10, 3):
    print(i, end=" ")
```

### enumerate()
> 인덱스와 값을 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">동시에</mark> 가져올 때 사용합니다.

```python
# TODO 5: names에 "Alice", "Bob", "Charlie"를 저장하고, 인덱스와 함께 순회하며 이름을 f-string으로 출력해봅시다.
names = ["Alice", "Bob", "Charlie"]
for index, name in enumerate(names):
    print(f"인덱스: {index}, 이름: {name}")
```

</br>

### zip()
> 여러 반복 가능 객체를 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">병렬로 순회</mark>합니다.

```python
# TODO 6: names에 "Alice", "Bob"를 저장하고, scores에 90, 85를 저장한 후, 병렬로 순회하며 이름과 점수를 f-string으로 출력해봅시다.
names = ["Alice", "Bob"]
scores = [90, 85]
for name, score in zip(names, scores):
    print(f"이름: {name}, 점수: {score}")
```
