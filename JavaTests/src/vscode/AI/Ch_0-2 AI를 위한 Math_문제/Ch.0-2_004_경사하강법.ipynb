```python
# TODO 0: 실습을 위해 아래 패키지를 import하고, diamonds 데이터셋을 준비해주세요.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

# diamonds 데이터셋 로드 (carat → price 예측)
df = sns.load_dataset("diamonds")
X = df[["carat"]].values
y = df["price"].values

# 표준화 (경사하강법 수렴을 위해 필수)
X_mean, X_std = X.mean(axis=0), X.std(axis=0)
y_mean, y_std = y.mean(), y.std()
X_scaled = (X - X_mean) / X_std
y_scaled = (y - y_mean) / y_std

# 바이어스 항 추가
X_b = np.c_[np.ones((len(X_scaled), 1)), X_scaled]

# Train/Valid 분할 (80:20) — 003_데이터 분할에서 학습한 내용
dataset_count = len(X_b)
shuffled_index = np.random.permutation(dataset_count)
cut_index = int(dataset_count * 0.8)
X_train, X_valid = X_b[shuffled_index[:cut_index]], X_b[shuffled_index[cut_index:]]
y_train, y_valid = y_scaled[shuffled_index[:cut_index]], y_scaled[shuffled_index[cut_index:]]

print(f"데이터: diamonds (carat → price, {len(X)}개)")
print(f"Train: {len(X_train)}개, Valid: {len(X_valid)}개")
```

# 경사하강법 (Gradient Descent)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">비용 함수의 기울기(Gradient) 반대 방향</mark>으로 파라미터를 반복 업데이트하여 최솟값을 찾는 알고리즘입니다.

$$\theta_{t+1} = \theta_t - \alpha \cdot \nabla J(\theta_t)$$

## 비용 함수 (Cost Function)
> 모델의 예측값과 실제값 사이의 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">오차를 수치화</mark>한 함수입니다.

### MSE (Mean Squared Error)

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

```python
# TODO 1: MSE를 계산하는 mean_squared_error 함수를 정의해봅시다.

def mean_squared_error(X, y, theta):
    data_count = len(y)
    predictions = X @ theta # y_hat -> 예측값
    
    cost = (1/data_count) *np.sum((y-predictions)**2)
    
    return cost
```

```python
# TODO 2: theta를 0으로 초기화 한 행렬을 생성해봅시다.

theta = np.zeros(X_train.shape[1])
print(theta)
```

```python
# TODO 3: mean_squared_error 함수를 이용하여 cost를 계산해봅시다.

cost = mean_squared_error(X_train, y_train, theta)
print(cost)
```

## 경사하강법 구현

> MSE를 $\theta$로 미분하여 그래디언트를 구한 뒤, 파라미터를 반복 업데이트합니다.

### 1단계: 예측값 정의

$$\hat{y} = X\theta$$

### 2단계: MSE를 행렬로 표현

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \frac{1}{n}(y - X\theta)^T(y - X\theta)$$

### 3단계: $\theta$로 미분 (연쇄 법칙)

> $(y - X\theta)^T(y - X\theta)$를 $\theta$로 미분하면, $(y - X\theta)$의 미분 $-X$와 제곱의 미분 $2$가 결합됩니다.

$$\nabla_\theta MSE = \frac{1}{n} \cdot 2 \cdot (-X)^T(y - X\theta) = \frac{2}{n} X^T(X\theta - y)$$


### 4단계: 파라미터 업데이트

$$\theta_{t+1} = \theta_t - \alpha \cdot \nabla_\theta MSE$$

</br>

```python
# TODO 4: 경사하강법을 위한 gradient_descent 함수를 정의해봅시다.

def gradient_descent(X, y, theta, learning_rate=0.01, iterations=500):

    # TODO 4-1: data_count에 데이터의 개수를 저장해봅시다.
    data_count = len(y)
    
    # TODO 4-2: 반복 횟수만큼 비용을 저장하기 위한 cost_list를 생성해봅시다.
    cost_list = []
    
    # TODO 4-3: iterations 만큼 반복을 위한 for 문을 작성합니다.
    for iteration in range(iterations):

        # TODO 4-3-1: gradient에 기울기를 저장해봅시다.
        gradient = (2/data_count) * X.T @ (X @ theta - y) 

        # TODO 4-3-2: theta를 업데이트해봅시다.
        theta = theta - learning_rate * gradient

        # TODO 4-3-3: mean_squared_error 함수를 이용하여 cost에 비용을 계산해봅시다.
        cost = mean_squared_error(X,y,theta)

        # TODO 4-3-4: cost_list에 해당 비용을 저장해봅시다.
        cost_list.append(cost)

        # TODO 4-3-5: 100회마다 cost를 출력해봅시다.
        if iteration % 100 == 0:
            print(iteration, cost)

    # TODO 4-4: theta와 cost_list를 반환해봅시다.
    return theta, cost_list
```

```python
# TODO 5: gradient_descent 함수를 이용하여 학습률은 0.01, 반복 횟수는 500으로 계산해봅시다.
theta_final, cost_list = gradient_descent(X_train, y_train, theta)

print(theta_final)
```

> 경사하강법은 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">대규모 데이터에서도 효율적</mark>이며, 대부분의 딥러닝 모델이 이 방법을 사용합니다.
